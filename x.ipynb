```python
import numpy as np

def ImplementGradientDescent(x, y, iterations, a, b):
    m = len(x)
    lr = 0.001
    for i in range(iterations):
        # Calculate the predictions
        y_preds = a + b * x
        
        # Calculate the errors
        error = y_preds - y
        
        # Calculate the cost (mean squared error)
        cost = (1/(2*m)) * np.sum(error**2)
        
        # Calculate gradients
        error_a = np.sum(error) / m
        error_b = np.sum(error * x) / m
        
        # Update parameters
        a -= lr * error_a
        b -= lr * error_b
        
        # Optionally print error for monitoring
        if i % 1 == 0:  # Adjusted to print every iteration since there are only 10 iterations
            print(f"After {i} iterations: a = {a}, b = {b}, cost = {cost}")

    print(f"After {iterations} iterations: a = {a}, b = {b}")

if __name__ == "__main__":
    x = []
    y = []
    data_length = 5
    iterations = 10000
    for i in range(data_length):
        x.append(i)
        y.append(i * 3 + 2)
    
    print(x)
    print(y)
    
    ImplementGradientDescent(np.array(x, dtype="int64"), np.array(y, dtype="int64"), iterations, 0, 0)
```
